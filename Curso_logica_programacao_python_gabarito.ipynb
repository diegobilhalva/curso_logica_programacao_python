### **1.8 Ducking Typing**

>_Se anda como um Pato e faz $quack$ como um pato, então é um pato_

image.png

A expressão divertida acima refere-se que caso você não se importe com o tipo de um objeto, basta satisfazer-se analisando se determinado objeto possui determinados métodos e atributos. Por exemplo, você pode verificar se determinado **objeto é iteravel** executando o código abaixo. Veja:

````python
def ehiteravel(obj):
  try:
    iter(obj)
    return True
  except TypeError: #não é iterável
    return False
````

Se você observar bem o código da função acima, irá notar que esta retornará $verdadeiro$ ($True$) se o objeto puder sofrer iterações. Do contrário irá retornar $Falso$ ($False$) conforme tratamento de erro. Experimente:

````python
a = 'strings são iteráveis'
ehiteravel(a)
````

````python
b = ['listas', 'são', 'iteráveis', (1, 2)]
ehiteravel(b)
````

````python
#inteiro não é iterável
c = 7
ehiteravel(c)
````

### **1.10 Operadores Binários e Comparações**

#### **1.10.1 Operadores Matemáticos usuais / Binários**: 

Os operadores matemáticos se comportam de maneira esperada e são eles abaixo relacionados:

| Operação 	| Descrição 	|
|---	|---	|
| $a$ $+$ $b$ 	| Soma o valor de $a$ ao valor de $b$. 	|
| $a$ $-$ $b$ 	| Subtrai o valor $b$ de $a$. 	|
| $a$ $*$ $b$ 	| Multiplica o valor de $a$ ao valor de $b$. 	|
| $a$ $/$ $b$ 	| Divide o valor de $a$ pelo valor de $b$. 	|
| $a$ $//$ $b$ 	| Divide o valor de $a$ pelo piso do valor de $b$, descartando qualquer resto fracionário. 	|
| $a$ $**$ $b$ 	| Submete o valor de $a$ à potência do valor de $b$. 	|
| $a$ $\%$ $b$ 	| Retorna o resto da divisão do numerador $a$ pelo denominador $b$. 	|

````python
a = 7
b = 4
````

````python
print("O resultado de 'a + b'  é: {0}".format(a + b))
print("O resultado de 'a - b'  é: {0}".format(a - b))
print("O resultado de 'a * b'  é: {0}".format(a * b))
print("O resultado de 'a / b'  é: {0}".format(a / b))
print("O resultado de 'a // b' é: {0}".format(a // b))
print("O resultado de 'a ** b' é: {0}".format(a ** b))
print("O resultado de 'a % b'  é: {0}".format(a % b))
````

#### **1.10.2 Operadores de Atribuição concomitantemente com os operadores Binários**:

Atribuição é quando passamos um valor (à direita da igualdade) para uma **variável** (à esquerda da igualdade, como o exemplo abaixo:
> $var$ = $'valor$ $qualquer'$ <br>(sentido da atribuição: **$var$** <font size='24'>&larr;</font> $'valor$ $qualquer'$)

| Operação 	| Descrição 	|
|---	|---	|
| $a$ $=$ $b$ 	| b é atribuido à variável a, fazendo com que ela torne-se b. 	|
| $a$ $+=$ $b$ 	| Atribui=se a soma ao valor de $a$, o valor de $b$. 	|
| $a$ $-=$ $b$ 	| Atribui-se a subtração do valor de $a$, o valor de $b$. 	|
| $a$ $*=$ $b$ 	| Atribui-se a multiplicação ao valor de $a$ pelo valor de $b$. 	|
| $a$ $/=$ $b$ 	| Atribui-se a divisão do valor de $a$ pelo valor de $b$. 	|
| $a$ $//=$ $b$ 	| Atribui-se a divisão do valor de $a$ pelo piso do valor de $b$, descartando qualquer resto fracionário. 	|
| $a$ $\%=$ $b$ 	| Atribui-se o resto da divisão do numerador $a$ pelo denominador $b$. 	|

````python
a = 7
b = 4
a += b 

print('O resultado da atribuição "a += b" é {0}'.format(a))
print('O valor de "a" é: {0}'.format(a))
print('O valor de "b" é: {0}'.format(b))
````

````python
a = 7
b = 4
a -= b 

print('O resultado da atribuição "a -= b" é {0}'.format(a))
print('O valor de "a" é: {0}'.format(a))
print('O valor de "b" é: {0}'.format(b))
````

Com os experimentos anteriores podemos comprovar uma propriedade interessante do **Python**. Diferente de outras linguagens em que quando se atribui o valor de uma variável $X$ qualquer a outro objeto $Y$, ao invés de $Y$ apenas assumir distintamente os valores de $X$, $Y$ acaba tornando-se $X$.

Entretanto, quando queremos comparar (saber) algo sobre uma varável (objeto), damos o nome de **comparação** através dos operadores listados na tabela abaixo:

#### **1.10.3 Comparações ou Operadores Relacionais**


| Operação 	| Descrição 	|
|---	|---	|
| $a$ $==$ $b$ 	| $True$ se a for igual a b 	|
| $a$ $!=$ $b$ 	| $True$ se a for diferente de b 	|
| $a$ $<$ $b;$ $a$ $<=$ $b$ 	| $True$ se a for menor que (menor ou igual a) b 	|
| $a$ $>$ $b;$ $a$ $>=$ $b$ 	| $True$ se a for maior que (maior ou igual a) b 	|
| $a$ $is$ $b$ 	| $True$ se a e b referenciarem o mesmo objeto Python 	|
| $a$ $is$ $not$ $b$ 	| $True$ se a e b referenciarem objetos Python diferentes 	|
| $a$ & $b$; $a$ $and$ $b$ 	| $True$ se tanto a quanto b forem $True$; para inteiros é a operação bit a bit (_bitwise_) $"and"$ 	|
| $a$ $|$ $b$; $a$ $or$ $b$ 	| $True$ se $a$ ou $b$ forem $True$; para inteiros, é a operação bit a bit $"or"$ 	|
| $a$ ^ $b$ 	| Para booleanos, $True$ se a ou b forem $True$ mas não ao mesmo tempo; para inteiros, é a operação bit a bit $exclusive-or$ 	|
| $a$ in $b$ 	| $True$ se $a$ é elemento de $b$ 	|
| $a$ not in $b$ 	| $True$ se $a$ não é elemento de $b$ 	|

**Enumerate**:

Imagina que você quer criar um dicionário (assunto que veremos em breve), para controlar os índices atuais dos elementos de uma _lista_ ou _tupla_. Uma maneira de se fazer é conforme abaixo:

````python
i = 0 #inteiro que utilizaremos como índice

mapping = {} #dicionário que armazenaremos a chave e o valor

lista = ['fulano', 'sicrano', 'beltrano']

for elemento in lista:
  print(elemento)
  mapping[i] = elemento
  i += 1

mapping
````

Como o uso deste tipo de recurso é bem comum, principalmente quando se trabalha com manipulação de dados, a comunidade de desenvolvedores do **Python** resolveram nos presentear com a *função embutida* $enumerate$. Veja:

````python
lista = ['fulano', 'sicrano', 'beltrano']

mapping = {}
for i, elemento in enumerate(lista):
  mapping[i] = elemento
mapping
````
