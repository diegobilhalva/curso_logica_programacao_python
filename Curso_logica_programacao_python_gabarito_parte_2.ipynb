### **2.2 Laços for**

Em **Python**, assim como na maioria das linguagens que possuem este mesmo laço, o utilizam para iterar por uma coleção, tais como uma "lista" ou uma "tupla", ou um iterador.

A sintaxe padrão segue o preceito abaixo:



```
for valor in colecao:
  #faz alguma coisa
```



````python
for i in range(10):
  print(i)
````

A instrução **break**:

````python
for i in range(10):
  if i > 5:
    print('a condição foi satisfeita')
    break; #encerra o laço do qual ele faz parte
  print(i)
````

Para casos em que existe **um laço dentro de outro laço**, note o comportamente do **break**:

````python
for i in range(14):
  for j in range(3):
    if i > 5:
      print('a condição foi satisfeita')
      break; #encerra o laço do qual ele faz parte
    print((i,j))
  print(i)
````

Mais a frente veremos sobre **desempacotamente de listas e tuplas**, mas já observa o quão interessante é usar este tipo de técnica. Veja:

````python
lista = [
         ('Diego',5)
         , ('João',3)
         , ('José', 6)
         , ('Maria', 9)
]

for nome, nota in lista:
  print('O aluno {0:} obteve grau {1:n}'.format(nome, nota))
````

### **2.3 Laços while**

Aqui, é importante levar em consideração que enquanto a condição for satisfeita, o laço continuará em execução. Então, muito cuidado neste tipo de implementação para não criar os famosos _loops infinitos_.

````python
x = 0
condicao = True

while condicao:
  if x > 5:
    condicao = False #break também funciona. Utilizá-lo ou controlar a 'condição' é uma decisão estratégica do programador.
  print(x)
  x += 1
````

````python
comando = ''
i = 0
while comando != 'sair':
  i += 1
  print('executado {0:n}'.format(i))
  comando = input('digite o comando "sair" para parar a execução:')
````

**#dicaTT / #dicadoBilly**: Existe um método while, perigoso é verdade, mas muito útil quando se está programando segundo um paradigma de eventos. O exemplo está listado abaixo.

````python
evento = False
while 1:
  if evento:
    break
  elif input('digite o comando "sair" para parar a execução:') == 'sair':
    print('o evento esperado aconteceu')
    evento = True
  else:
    print('a execução continua até o evento esperado acontecer')
````

Uma _expressão ternária_ em Python permite combinar um block $if-else$ que gera um valor em uma única linha ou expressão conforme a sintaxe abaixo:

$value$ $=$ $true-expr$ $if$ $condition$ $else$ $false-expr$

#### E5. ⏰(5min) Crie uma sequência de Fibonnaci com 20 termos.

````python
elemento0 = 0
elemento1 = 1
print('A sequencia de Fibonacci é: {0} {1}'.format(elemento0, elemento1), end = ' ')
for i in range(0,18):
  elemento2 = elemento1 + elemento0
  elemento0 = elemento1
  elemento1 = elemento2
  i += 1
  print(elemento2, end = ' ')
````

#### E6. ⏰(5min) Escreva todos os números primos existentes entre 2 e 30.

````python
for numero in range(2, 30):
  contador_divisoes_com_resto_zero = 0
  
  for divisor in range(1, numero + 1):
    if numero % divisor == 0:
      contador_divisoes_com_resto_zero += 1

  if contador_divisoes_com_resto_zero == 2:
    print(numero, end=' ')
  
````

#### E7. ⏰(15min) Crie um **sistema em Python** que **calcule o faturamento líquido** (valor bruto menos impostos) de chocolates em barra, de uma fábrica de chocolates, onde o analista comercial **informe a quantidade vendida, o valor unitário e o imposto em % do referido produto**, entretanto se a quantidade vendida for menor que 50 barras, o imposto devido deve ser acrescido de 5%, não podendo ultrapassar 20% do imposto total: 

````python
qtty = int(input('Informe a quantidade: '))
valor = float(input('valor unitário: '))
imposto = float(input('informe o imposto em valor percentual SEM %: '))/100

if qtty < 50:
  imposto += .05

if imposto > .2:
  imposto = .2

faturamento = (qtty*valor)*(1-imposto)
print('O faturamento foi de R${0:.2f}'.format(faturamento))
print('O imposto aplicado foi de {0:.2%}'.format(imposto))
````

#### E8. ⏰(30min) Crie um **sistema em Python** que **calcule o troco devido** de um montante pago em relação ao valor devido. Os requisitos são os abaixo definidos: 

1.   O sistema deve **questionar o valor da compra em reais**;
2.   O sistema deve **questionar o valor pago ao caixa em reais**;
3.   O sistema deve **calcular o troco a receber**.
4.   O sistema deve **perguntar ao pagador** se ele **quer receber seu troco no máximo múltiplo de 5 menor que o troco a ser pago e doar o restante para a caridade**, informando os valores do troco a receber e a quantidade a ser doada para a caridade.
  * **Exemplo**: Vamos supor que o **valor da compra foi de R\$13,25**. O pagador dá uma nota de **R\$20,00**. O **troco devido será de R\$6,75**, de forma que o **sistema deve perguntar se o pagador quer doar R\$1,75** para a caridade e **receber R\$5,00 de troco**.
5. **Caso o pagador não queira doar**, devolver o troco a ser pago.
6. **Caso o montante pago seja menor que o valor da compra**, informar ao pagador o quanto falta para completar o pagamento devido.
7. **Fim do processo**

````python
valor_da_compra = float(input('Qual o valor da compra? '))
qtty_recebida = float(input('Qual o valor recebido? '))

troco = qtty_recebida - valor_da_compra
print('O troco devido é de R${0:.02f}'.format(troco))


caridade = round(troco%5,2)

if caridade > 0 and troco > 0:
  quer_doar_para_a_caridade = input('Responda "sim" se você deseja doar R${0} para a caridade? '.format(caridade%5))

  if quer_doar_para_a_caridade == 'sim':
    print('O(a) senhor(a) está doando R${0:.02f} para a caridade e aqui está o seu troco de R${1:.02f}. '.format(troco%5, troco-(troco%5)))
  else:
    print('Aqui está seu troco de R${0:.02f} e obrigado pela compra.'.format(troco))
elif troco < 0:
  print('Está faltando R${0:.2f}.'.format(troco))
else:
  print('Obrigado pela compra.'.format(troco))
````
